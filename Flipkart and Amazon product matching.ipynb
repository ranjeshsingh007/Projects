```python
 # using the "all-MiniLM-L6-v2" Sentence Transformer model for calculating semantic similarity 
model = SentenceTransformer('all-MiniLM-L6-v2')

#embedding for both Amazon and Flipkart
amazon_embs = model.encode(amazon_dataset.product_name,convert_to_tensor=True) 
flipkart_embs = model.encode(flipkart_dataset.product_name,convert_to_tensor=True) 

cosine_sim = util.cos_sim(amazon_embs,flipkart_embs) # cosine similaritise
```

```python
# Storing the index of a similar Flipkart product  to each Amazon product 
similar_prod_indexes = []

for n in range(flipkart_dataset.shape[0]):
    max_score_idxs = torch.topk(cosine_sim[n],k=5,largest=True,sorted=True).indices
    for index in max_score_idxs:
        if index != n:
            similar_prod_indexes.append(index)
            break
```

```python
similar_prod_indexes = [x.item() for x in similar_prod_indexes]
```

```python
amazon_dataset.product_name[1444], flipkart_dataset.product_name[similar_prod_indexes[1444]]
```

```python
# Extracting the details such as retail price and discounted price for both Flipkart and Amazon similar products
flipkart_prod_data = []
amazon_prod_data = []

for idx, prod_name in enumerate(amazon_dataset.product_name):
    flipkart_prod_data.append(flipkart_dataset.iloc[similar_prod_indexes[idx]])
    amazon_prod_data.append(amazon_dataset.iloc[idx])
```

```python
flipkart_prod_data = pd.DataFrame(flipkart_prod_data)
flipkart_prod_data.columns = ['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart']
amazon_prod_data = pd.DataFrame(amazon_prod_data)
amazon_prod_data.columns = ['Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']
```

#### amazon retail price and discounted price is not in form as flipkart price so we need to change it into same as float64

```python
amazon_prod_data['Retail Price in Amazon'] = amazon_prod_data['Retail Price in Amazon'].astype(np.float64)
amazon_prod_data['Discounted Price in Amazon'] = amazon_prod_data['Discounted Price in Amazon'].astype(np.float64)
```

## for concating flipkart and amazon dataset we need to reset flipkart indexes

```python
flipkart_prod_final_data = flipkart_prod_data.copy() # copy for future refernece
flipkart_prod_final_data = flipkart_prod_final_data.reset_index(drop=True)
flipkart_prod_final_data.head()
```
